```python
import requests
from bs4 import BeautifulSoup
```

```python
from ddgs import DDGS
```

```python
def search_sector(sector: str):
    query = f"India {sector} sector trade opportunities 2024"
    results = DDGS().text(query, max_results=5)
    return results  # list of {title, href, body}
```

```python
def extract_text_from_url(url: str) -> str:
    try:
        response = requests.get(url, timeout=10, headers={
            "User-Agent": "Mozilla/5.0"
        })
        soup = BeautifulSoup(response.text, "html.parser")

        for tag in soup(["script", "style", "nav", 
                          "footer", "header"]):
            tag.decompose()

        text = soup.get_text(separator=" ", strip=True)
        lines = [line.strip() for line in text.splitlines() if line.strip()]
        return " ".join(lines)[:2000]

    except:
        return ""
```

```python

def collect_data(sector: str) -> str:
    search_results = search_sector(sector)
    
    all_text = ""
    
    for result in search_results:
        # Use the snippet directly from DuckDuckGo
        all_text += f"\nSource: {result['title']}\n"
        all_text += f"Summary: {result['body']}\n"
        
    
    return all_text
```
